```python
# LOOKING AT THE DATA FROM THE POINT OF VIEW OF GENDER
```

```python
G=sns.countplot(x='Gender',data=df)
for bars in G.containers:
    G.bar_label(bars)
```

```python
# The above graph shows the number of men vs women who have been involved in the purchase transactions.
# FEMALES HAVE PURCHASED MORE IN COMPARISION TO MEN.
```

```python
sales_By_Gender=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_By_Gender)
```

```python
# The purchasing power of women is more than that of men.
```

```python
A=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in A.containers:
    A.bar_label(bars)
```

```python
sales_By_Age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_By_Age)
```

```python
# ANALYSIS BY STATES
```

```python
Sales_By_State=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=Sales_By_State)
```

```python
# Here are the top states in terms of sales Uttar Pradesh tops te list followed by Maharashtra and kerela is at the bottom
```

```python
Sales_By_State=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(8)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=Sales_By_State)
```
